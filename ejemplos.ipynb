```python
def process_sentence(sentence):
    sentence = sentence.lower()
    abiertas = sentence.count("a")+sentence.count("e")+sentence.count("o")
    cerradas = sentence.count("i")+sentence.count("u")
    return abiertas, cerradas, len(sentence)

v_abiertas, v_cerradas, longitud = process_sentence("Python 2026")
print (v_abiertas, v_cerradas, longitud)
```

```python
def saludo(nombre, mensaje="Hola"):
    print(mensaje, nombre)

saludo("Nair")          # usa "Hola"
saludo("Nair", "Chau")  # sobrescribe
```

```python
def my_args(*args):
    for v in args:
        print(v, type(v))

my_args(1, "hola", [1,2])
```

```python
def my_args1(**kwargs):
    for k, v in kwargs.items():
        print(k, v)

my_args1(band1="Nirvana", band2="AC/DC")
```
