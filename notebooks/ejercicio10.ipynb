```python
#Ejercicio 2: Duración de una playlist
```

```python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]
```

```python
total_seconds=0

cancion_larga = None
cancion_corta = None
max_segundos = -1
min_segundos = float('inf')

for song in playlist:
    #itero el diccionario, y de cada canción pregunto la duración y lo paso como parámetro para la función
    segundos = duration_to_seconds (song['duration'])
    #acumulo el total de segundos que va devolviendo la función
    total_seconds+= segundos

    if segundos > max_segundos:
        #compara y busca el máximo
        max_segundos=segundos
        #guarda el diccionario de la canción
        cancion_larga= song
    if segundos < min_segundos: #hace lo mismo pero con los mínimos
        min_segundos=segundos
        cancion_corta=song

print (f"Procesamiento completado")
```

```python
#convertimos los segundos totales a minutos y segundos
total_minutes = total_seconds // 60
remaining_seconds= total_seconds%60

print (f"Duración total: {total_minutes}m {remaining_seconds}s")
#accedemos al diccionario guardado en la variable de máx/min
print(f"Canción más larga: \"{cancion_larga['title']}\" ({cancion_larga['duration']})")
print (f"Canción más corta: \"{cancion_corta['title']}\"({cancion_corta['duration']})")
```

```python
#Ejercicio 6: Análisis de hashtags

posts = [
    "Arrancando el lunes con energía #Motivación #NuevaSemana",
    "Terminé mi primer proyecto en Python #Python #Programación #OrgullosoDeMi",
    "No puedo creer el final de la serie #SinSpoilers #SerieAdicta",
    "Nuevo video en el canal sobre #InteligenciaArtificial y #Python",
    "Entrenamiento de hoy completado #Fitness #Motivación #NoPainNoGain",
    "Leyendo sobre #InteligenciaArtificial y el futuro del trabajo #Tecnología",
    "Arranqué a estudiar #Programación por mi cuenta #Python #Autodidacta",
    "Finde de lluvia, maratón de series #SerieAdicta #Relax",
    "Workshop de #InteligenciaArtificial en la universidad #Tecnología #Programación"
]

#analizar
trending, total_unicos, contador_completo = analizar_hashtags(posts)

#mostrar resultados
print ("Hashtags trending (más de una aparición): ")
#ordena (sorted)los elementos (items) de la lista trending usando como criterio el número (key=lambda x:x[1]), poniendo 1ero el que tiene nro más grande (reverse=True)
#x[0] -> sería el nombre del hashtag (la palabra)
#x[1] -> es la cantidad (el nro)
for hashtag, cantidad in sorted(trending.items(), key = lambda x: x[1], reverse= True): 
    print (f"{hashtag}: {cantidad}")

print (f"\nTotal de hashtags únicos: {total_unicos}")
```
