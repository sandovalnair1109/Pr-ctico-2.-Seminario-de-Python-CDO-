```python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]
```

```python
def duration_to_seconds(duration_str):
    #le paso como parámetro la duración de cada canción
    """
    Convierte string "m:ss" a segundos totales.
    Ejemplo: "5:55" --> 5*60 +55 = 355 segundos
    """

    minutes, second = duration_str.split(":") 
    return int(minutes)*60 + int(second) #devuelve el formato que se pide

    print (duration_to_seconds("5:55"))

total_seconds=0

cancion_larga = None
cancion_corta = None
max_segundos = -1
min_segundos = float('inf')

for song in playlist:
    #itero el diccionario, y de cada canción pregunto la duración y lo paso como parámetro para la función
    segundos = duration_to_seconds (song['duration'])
    #acumulo el total de segundos que va devolviendo la función
    total_seconds+= segundos

    if segundos > max_segundos:
        #compara y busca el máximo
        max_segundos=segundos
        #guarda el diccionario de la canción
        cancion_larga= song
    if segundos < min_segundos: #hace lo mismo pero con los mínimos
        min_segundos=segundos
        cancion_corta=song

print (f"Procesamiento completado")
```

```python

#convertimos los segundos totales a minutos y segundos
total_minutes = total_seconds // 60
remaining_seconds= total_seconds%60

print (f"Duración total: {total_minutes}m {remaining_seconds}s")
#accedemos al diccionario guardado en la variable de máx/min
print(f"Canción más larga: \"{cancion_larga['title']}\" ({cancion_larga['duration']})")
print (f"Canción más corta: \"{cancion_corta['title']}\"({cancion_corta['duration']})")
```

```python
def censurar_texto(texto, palabras_prohibidas):
    """
    Reemplaza cada palabra prohibida por astericos (*).
    Mantiene el mismo número de caracteres.

    """
    texto_censurado= texto
    for palabra in palabras_prohibidas:
            #nos aseguramos que los asteriscos tengan la misma longitud
            astericos="*" * len(palabra)
            #llamo a la función que reemplaza las palabras por los asteriscos en el texto
            texto_censurado= reemplazar_ignore_case(texto_censurado, palabra, astericos)
            #devuelvo el texto ya censurado
            return texto_censurado

#al ser una función dentro de una iteración, va a ser llamada hasta que se terminen las palabras
def reemplazar_ignore_case(texto, objetivo, reemplazo):
    """ 
    Reemplaza 'objetivo' por 'reemplazo' en 'texto'
    """
    import re
    #re.IGNORECASE hace que no importe si es mayus/minus
    #re.escape ignora los caracteres especiales
    patron= re.compile (re.escape(objetivo),re.IGNORECASE)
    #sustituimos con .sub()
    return patron.sub(reemplazo, texto)
```
